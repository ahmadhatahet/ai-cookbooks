```python
from pprint import pprint
from pathlib import Path
from dotenv import load_dotenv

# load openai api key
load_dotenv(Path().cwd().parent / '.env')
```

```python
from langchain.chat_models import ChatOpenAI
llm = ChatOpenAI(model="gpt-3.5-turbo-16k", temperature=0, max_tokens=None)
```

```python
from langchain.agents import create_sql_agent
from langchain.agents.agent_toolkits import SQLDatabaseToolkit
from langchain.sql_database import SQLDatabase

DB_NAME='AdventureWorksDW2016'
DB_HOST='localhost\\SQLEXPRESS'
DB_USER='testuser'
DB_PASSWORD='Aa12345'

conn_uri = f'mssql+pyodbc://{DB_USER}:{DB_PASSWORD}@{DB_HOST}/{DB_NAME}'
conn_uri += '?driver=ODBC+Driver+17+for+SQL+Server'

db = SQLDatabase.from_uri(conn_uri)

toolkit = SQLDatabaseToolkit(db=db, llm=llm)

executer = create_sql_agent(
    llm=llm,
    toolkit=toolkit,
    verbose=True
)
```

```python
executer.run("What are the top three sales territories?")
```

```python
executer.run("What is the top sold product for each of the top three sales territories and how much was the sales?")
```
